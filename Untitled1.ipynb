```python
import pandas as pd
dataframe = pd.read_csv('Top Cars Database - Africa.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data(file_path):
    dataframe = pd.read_csv(file_path)
    return dataframe

def data_analysis(dataframe):
    # Average price per brand
    avg_price_per_brand = dataframe.groupby('Brand')['Price (USD)'].mean()
    
    # Most profitable car model
    most_profitable = dataframe.loc[dataframe['Profit (USD)'].idxmax(), ['Brand', 'Model', 'Profit (USD)']]
    
    # Count of cars by segment
    cars_by_segment = dataframe['Segment'].value_counts()
    
    return avg_price_per_brand, most_profitable, cars_by_segment

def visualize_data(dataframe):
    plt.figure(figsize=(12, 5))
    
    # Bar chart: Average price per brand
    avg_price_per_brand = dataframe.groupby('Brand')['Price (USD)'].mean().sort_values()
    avg_price_per_brand.plot(kind='bar', color='skyblue')
    plt.title('Average Car Price per Brand')
    plt.ylabel('Price (USD)')
    plt.xticks(rotation=45)
    plt.show()
    
    # Pie chart: Car distribution by segment
    plt.figure(figsize=(8, 8))
    dataframe['Segment'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
    plt.title('Car Distribution by Segment')
    plt.ylabel('')
    plt.show()
    
    # Scatter plot: Price vs. Profit
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=dataframe['Price (USD)'], y=dataframe['Profit (USD)'], hue=dataframe['Brand'], palette='tab10', s=100)
    plt.title('Price vs. Profit')
    plt.xlabel('Price (USD)')
    plt.ylabel('Profit (USD)')
    plt.show()

def filter_cars(dataframe, price_range=None, fuel_type=None, segment=None):
    filtered_dataframe = dataframe.copy()
    
    if price_range:
        filtered_dataframe = filtered_dataframe[(filtered_dataframe['Price (USD)'] >= price_range[0]) & (filtered_dataframe['Price (USD)'] <= price_range[1])]
    
    if fuel_type:
        filtered_dataframe = filtered_dataframe[filtered_dataframe['Fuel'] == fuel_type]
    
    if segment:
        filtered_dataframe = filtered_dataframe[filtered_dataframe['Segment'] == segment]
    
    return filtered_dataframe

# Load data
data_file = "Top Cars Database - Africa.csv"
dataframe = load_data(data_file)

# Perform analysis
avg_price, most_profitable, segment_count = data_analysis(dataframe)
print("Average Price per Brand:\n", avg_price)
print("Most Profitable Car:\n", most_profitable)
print("Car Count by Segment:\n", segment_count)

# Visualization
visualize_data(dataframe)

# Example Filtering
filtered_cars = filter_cars(dataframe, price_range=(15000, 35000), fuel_type='Petrol', segment='Sedan')
print("Filtered Cars:\n", filtered_cars)
```
